```python
import pandas as pd
import requests

# --- Distance and Duration Function ---
def get_distance(frm, to):
    url = f"https://router.project-osrm.org/route/v1/driving/{frm[1]},{frm[0]};{to[1]},{to[0]}?overview=false"
    response = requests.get(url)
    data = response.json()
    distance_km = data['routes'][0]['distance'] / 1000
    duration_h = data['routes'][0]['duration'] / 3600
    return distance_km, duration_h

# --- Extract unique centers and clients ---
unique_dcs = df_day1['IDCentro'].unique()
unique_clients = merged_data[['CLIENT #', 'LATITUDE', 'LENGTH']].drop_duplicates().set_index('CLIENT #')

# --- Build DC coordinate dictionary from matching_centres ---
dc_coords = {
    row['CD']: (row['Latitud'], row['Longitud'])
    for _, row in matching_centres.iterrows()
    if row['CD'] in unique_dcs
}

# --- Initialize new empty DataFrame for distances ---
distance_df = pd.DataFrame(index=unique_dcs, columns=unique_clients.index)

# --- Loop over all DCs and all clients ---
for dc_id in unique_dcs:
    if dc_id not in dc_coords:
        print(f"⚠️ Skipping {dc_id}: no coordinates found in matching_centres.")
        continue

    dc_loc = dc_coords[dc_id]

    for client_id, row in unique_clients.iterrows():
        client_loc = (row['LATITUDE'], row['LENGTH'])

        try:
            distance_km, _ = get_distance(dc_loc, client_loc)
            distance_df.at[dc_id, client_id] = distance_km
        except Exception as e:
            print(f"❌ Error: DC {dc_id} → Client {client_id}: {e}")
            distance_df.at[dc_id, client_id] = None

# Convert all distances to float
distance_df = distance_df.astype(float)

# ✅ You now have a full distance matrix (DCs x Clients)
```

```python
# A. Demand per client
demand_dict = merged_data.groupby('CLIENT #')['Ordered_vol'].sum().to_dict()

# B. Capacity per facility (use real values or set high if unknown)
capacity_dict = merged_data.groupby('IDCentro')['Ordered_vol'].sum().to_dict()
```

```python
from ortools.linear_solver import pywraplp

solver = pywraplp.Solver.CreateSolver('SCIP')

clients = list(distance_df.columns)
facilities = list(distance_df.index)

# --- Decision variables ---
x = {(dc, cl): solver.IntVar(0, 1, f'x_{dc}_{cl}') for dc in facilities for cl in clients}
y = {dc: solver.IntVar(0, 1, f'y_{dc}') for dc in facilities}

# --- Objective: minimize transport cost ---
solver.Minimize(
    solver.Sum(x[dc, cl] * distance_df.loc[dc, cl] * demand_dict[cl]
               for dc in facilities for cl in clients)
)

# --- Constraints ---

# 1. Every client assigned to one facility
for cl in clients:
    solver.Add(solver.Sum(x[dc, cl] for dc in facilities) == 1)

# 2. Assign to open facilities only
for dc in facilities:
    for cl in clients:
        solver.Add(x[dc, cl] <= y[dc])

# 3. Respect capacity
for dc in facilities:
    solver.Add(
        solver.Sum(x[dc, cl] * demand_dict[cl] for cl in clients) <= capacity_dict[dc]
    )

# --- Solve ---
status = solver.Solve()

# --- Results ---
if status == pywraplp.Solver.OPTIMAL:
    print(f"✅ Total Cost: {solver.Objective().Value():,.2f}")
    print("\n📍 Open Facilities:")
    for dc in facilities:
        if y[dc].solution_value() > 0.5:
            print(f" - {dc}")

    print("\n📦 Client Assignments:")
    for cl in clients:
        assigned_dc = [dc for dc in facilities if x[dc, cl].solution_value() > 0.5][0]
        print(f"Client {cl} → DC {assigned_dc}")
else:
    print("❌ No optimal solution found.")
```
